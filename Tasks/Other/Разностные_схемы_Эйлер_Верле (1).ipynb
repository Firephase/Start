```python
#Разностная схема для решения диффура y' = func(y,x)

def solution(func,x,y0,step):
  #problem : dydx = func(y,x)
  #solve it by producing the scheme
  solution = [0 for i in x]
  solution[0] = y0
  for i in range(1,len(x)):
    solution[i] = solution[i-1] + step*func(solution[i-1],x[i-1])
  return solution
```

```python
#метод Верле разностная схема
def solution_verle(dydx,x,y0,step):
  #problem : der(y) = func(y,x)
  #func
  #solve it by producing the scheme
  solution = [0 for i in x]
  solution[0] = y0
  solution[1] = solution[0] + step**2 * dydx(solution[0],x[0])
  for i in range(1,len(x)):
    solution[i] = 2 * solution[i-1] - solution[i-2] + step * step * dydx(solution[i-1],x[i-1])
  return solution
```
